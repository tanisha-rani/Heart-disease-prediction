### Prepare Data

```python
dataset = dataset.copy()
dataset['HeartDisease'] = dataset['HeartDisease'].map({'Yes': 1, 'No': 0})
```

```python
numeric_cols = ['BMI','PhysicalHealth','MentalHealth','SleepTime']
cat_cols = ['Smoking','AlcoholDrinking','Stroke','DiffWalking','Sex','AgeCategory','Race',
            'Diabetic','PhysicalActivity','GenHealth','Asthma','KidneyDisease','SkinCancer']
```

## Split Dataset

```python
from sklearn.model_selection import train_test_split

X = dataset.drop(columns=['HeartDisease'])
y = dataset['HeartDisease']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)
```

 ## Preprocessing Pipelines

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

# Numeric Transformer â†’ Handles missing values + scales numbers
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),     
    ('scaler', StandardScaler())                       
])

# Categorical Transformer â†’ Handles missing values + encodes categories
cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),  
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))  
])

# ðŸ§© Step 3: Combine both into one preprocessor
preprocessor = ColumnTransformer([
    ('num', num_transformer, numeric_cols),
    ('cat', cat_transformer, cat_cols)
])
```

```python
dataset["Sex"].unique()
```

```python
X_train_prepared = preprocessor.fit_transform(X_train)
X_test_prepared  = preprocessor.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb

# Define both models
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)

lgbm = lgb.LGBMClassifier(
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    random_state=42
)
```

```python
rf.fit(X_train_prepared, y_train)
lgbm.fit(X_train_prepared, y_train)
```

```python
y_pred_rf = rf.predict(X_test_prepared)
y_pred_lgbm = lgbm.predict(X_test_prepared)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print("ðŸ”¹ Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
print("ðŸ”¹ LightGBM Accuracy:", accuracy_score(y_test, y_pred_lgbm))

print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("\nLightGBM Classification Report:\n", classification_report(y_test, y_pred_lgbm))

print("\nRandom Forest Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("\nLightGBM Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lgbm))
```

### Combine preprocessing and model into a single pipeline

```python
from sklearn.pipeline import Pipeline

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', lgb.LGBMClassifier(n_estimators=500, learning_rate=0.05, random_state=42))
])

model_pipeline.fit(X_train, y_train)
y_pred = model_pipeline.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

### Finalize & Save Your Model + Preprocessor

```python
import joblib

# Save both LightGBM model and preprocessor
joblib.dump(lgbm, "heart_disease_model.pkl")
joblib.dump(preprocessor, "preprocessor.pkl")

print("âœ… Model and preprocessor saved successfully!")
```
